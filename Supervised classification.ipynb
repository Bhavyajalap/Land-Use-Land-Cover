```python
# Define Landsat 8 collection
l8_collection = ee.ImageCollection('LANDSAT/LC08/C02/T1')

# Filter Landsat 8 collection by date
l8rw = l8_collection.filterDate('2020-01-01', '2020-12-31')

#countries
countries = ee.FeatureCollection('USDOS/LSIB_SIMPLE/2017')
roi =   countries.filter(ee.Filter.eq('country_na', 'India'))
print(type(roi))
m.addLayer(roi,{},'India',False)
m.centerObject(roi,4)

#Make a cloud free composite Landsat 8
image = ee.Algorithms.Landsat.simpleComposite(**{
            'collection': l8rw.filterDate('2020-01-1', '2020-12-31'),
            'asFloat': True
})

TrueColorVis = {'min':0, 'max': 0.3, 'bands':['B4','B3','B2']}
m.addLayer(image.clip(roi), TrueColorVis, 'composite')

#create training
label = "Class"
bands = ['B1','B2','B3','B4','B5','B7']
input_1 = image.select(bands)

# 1=Urban; 2=Crop; 3=Forest; 4= Barren; 5=water
# overlay the points on the landsat image
trainImage = input_1.sampleRegions(**{
  'collection' :training,
  'properties' : [label],
  'scale': 30
})
print(trainImage)

trainingData = trainImage.randomColumn()
trainSet = trainingData.filter(ee.Filter.lessThan('random',0.8))
testSet = trainingData.filter(ee.Filter.greaterThanOrEquals('random',0.8))

#Classification Model (cart)

classifier = ee.Classifier.smileCart().train(trainSet,label,bands)

classified = input_1.classify(classifier)

landsatPlatter = [
  '#ff5769', # Red - urban
  '#35d355', # Green - crop
  '#907119', # Brown - forest
  '#fffa2b',  # Yellow - barren
  '#1337d6', # Blue - water
  ]

m.addLayer(classified.clip(roi), {'platte': landsatPlatter, 'min':1, 'max':5}, 'classification')
m
```
